```python
def reading(f_name):

    TS = [[],[],[]]
    FS = [[],[],[]]
    PS = [[],[],[],[]]
    
    f = open(f_name, 'r')
    
    for line in f:
        if 'Grasp' in line:
            g = int(line[6:]) - 1
            for i in range(3):
                TS[i].append([])
                FS[i].append([])
                PS[i].append([])
            PS[3].append([])
        else:
            num = int(line[2])
            content = line[4:]
            if 'TS' in line:
                TS[num][g].append(list(eval(content))[2])
            if 'FS' in line:
                FS[num][g].append(list(eval(content))[2])
            if 'PS' in line:
                PS[num][g].append(list(eval(content))[2])
                
    f.close()
    return(TS, FS, PS)
```

```python
import matplotlib.pyplot as plt

f_name = '.\Data\Cuboid_2A1_0.txt'
TS_C_0, FS_C_0, PS_C_0 = reading(f_name)
```

```python
import random

plt.figure(figsize = (25,5))

rd = random.randint(1,199)

plt.subplot(1, 3, 1)
for i in range(3):
    plt.plot((TS_C_0[i][rd]), label = 'Thumb sensor ' + str(i))
plt.title('Readout of thumb sensors')
plt.xlabel('Time')
plt.ylabel('Readout')
plt.legend()

    
plt.subplot(1, 3, 2)
for i in range(3):
    plt.plot(FS_C_0[i][rd], label = 'Finger pulp sensor ' + str(i))
plt.title('Readout of finger pulp sensors')
plt.xlabel('Time')
plt.ylabel('Readout')
plt.legend()
    
plt.subplot(1, 3, 3)
for i in range(4):
    plt.plot((PS_C_0[i][rd]), label = 'Proximity sensor ' + str(i))
plt.title('Readout of proximity sensors')
plt.xlabel('Time')
plt.ylabel('Readout')
plt.legend()

plt.show()
```
